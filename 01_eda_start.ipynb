```python
df_da = df[df["title"].str.contains("Data Analyst", case=False, na=False)].copy()
```

```python
df_da["description_clean"] = df_da["description"].str.lower()

df_da["python"] = df_da["description_clean"].str.contains("python", na=False)
df_da["sql"] = df_da["description_clean"].str.contains("sql", na=False)
df_da["excel"] = df_da["description_clean"].str.contains("excel", na=False)
df_da["tableau"] = df_da["description_clean"].str.contains("tableau", na=False)
df_da["powerbi"] = df_da["description_clean"].str.contains("power bi", na=False)

df_da[["python","sql","excel","tableau","powerbi"]].sum()
```

### gráfico de skills

```python
import matplotlib.pyplot as plt

skills = df_da[["python","sql","excel","tableau","powerbi"]].sum()

skills.sort_values().plot(kind="barh", color="steelblue")
plt.title("Most In-Demand Data Analyst Skills")
plt.xlabel("Number of Job Postings")
plt.show()
```
